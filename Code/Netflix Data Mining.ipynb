## Data Modeling

We import the NLTK library to optimize the words present in the variables using stop work removal and stemming to index the words. 

Stop words help us identify the high frequency words that have little semantic weight and hence do not help with the retrieval process. Therefore, these words are dropped from the index. 

Stemming removes the ends of the words that help in removing the derivational affixes. It is a word reduction procedure that achieves to obtain the real word of meaning which could help with further analysis.

```python
import nltk

nltk.download('punkt')
nltk.download('stopwords')
```

```python
import nltk, re
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize

# Let's get a list of stop words from the NLTK library
stop_words = ["a", "an", "the", "this", "that", "is", "it", "to", "and"]

stemmer = SnowballStemmer('english')


def process_text(texts): 
    final_text_list=[]
    for sent in texts:
        filtered_sentence=[]
        
        #sent = sent.lower() # Lowercase 
        sent = sent.strip() # Remove leading/trailing whitespace
        sent = re.sub('\s+', ' ', sent) # Remove extra space and tabs
        sent = re.compile('<.*?>').sub('', sent) # Remove HTML tags/markups:
        
        for w in word_tokenize(sent):
            # We are applying some custom filtering here, feel free to try different things
            # Check if it is not numeric and its length>2 and not in stop words
            if(not w.isnumeric()) and (len(w)>2) and (w not in stop_words):  
                # Stem and add to filtered list
                filtered_sentence.append(stemmer.stem(w))
        final_string = " ".join(filtered_sentence) #final string of cleaned words
 
        final_text_list.append(final_string)
    
    return final_text_list
```

We identify whitespaces, tabs, html tags and markups as stop words and eliminate them from the text. Also, the Snowball Stemmer is used which is a small string processing language that helps in the process of Information Retrieval.

```python
print("Pre-processing the reviewText field")
data["description"] = process_text(data["description"].tolist()) 
```

```python
from sklearn.model_selection import train_test_split

train_data, val_data = train_test_split(data, test_size=0.1, shuffle=True, random_state=23)
```

Here, we split the dataset into train and validation sets to test the model's performance.

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

# Initialize the binary count vectorizer
tfidf_vectorizer = CountVectorizer(binary=True,
                                   max_features=50 # Limit the vocabulary size
                                  )
# Fit and transform
text_vectors = tfidf_vectorizer.fit_transform(train_data["description"].tolist())

# Only transform
X_val_text_vectors = tfidf_vectorizer.transform(val_data["description"].tolist())

#print tdidf vocabulary
print(tfidf_vectorizer.vocabulary_)
```

The above implementation is known as the counting bag of words. Basically, a count vectorizer returns the number of the frequency to index the vocabulary whereas ter frequency - inverse document frequency considers the overall documents' count of words. 

In other words, the function returns how many times a word is present int he document. For the high term frequency on the low document frequency of the term in the entire collection. Hence, the overall document's weightage is considered in processing a word helping us identify how often these words are encountered in the documents. 

```python
from sklearn.cluster import KMeans

kvec = []
#true_k = 3
for i in range (1, 30):
    model = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0, verbose=True)
    model.fit(text_vectors)
    kvec.append(model.inertia_)
```

We implement the KMeans clustering technique to group our observations into k-clusters that deal with the most frequent words calculating the most optimum number. After removing the stop words, stemming the text and tokenizing the words, we apply k-means clustering using the euclidean distance.  

```python
plt.plot(range(1, 100), kvec)
plt.title('The elbow method')
plt.xlabel('no. of clusters')
plt.ylabel('kvec')
plt.show()
```

Here, the elbow method determines the optimal number of clusters to use. We plot the elbw curve to understand this.

```python
true_k = 10
model = KMeans(n_clusters=true_k, init='k-means++', n_init=1)
model.fit(text_vectors)
```

```python
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = tfidf_vectorizer.get_feature_names()
```

```python
for i in range(true_k):
 print("Cluster %d:" %i),
 for ind in order_centroids[i, :10]:
     print("%s" % terms[ind])
```
